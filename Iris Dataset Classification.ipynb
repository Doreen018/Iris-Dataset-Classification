```python
from sklearn.datasets import load_iris
iris_dataset=load_iris()
print(iris_dataset.keys())
```

```python
#Splitting the Dataset
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test =train_test_split(
iris_dataset['data'], iris_dataset['target'], test_size=0.3, random_state=0) #Test size is 30% (0.3) train size is 0.7 (70%)
```

```python
#inspecting data using visualization by using pandas
#Creating dataframe from data in X_train
#label the columns using the strings in iris_dataset.feature_names
import pandas as pd
iris_dataframe=pd.DataFrame(X_train, columns=iris_dataset.feature_names)
#Creating a scatter matrix from the dataframe, color by y_train
pd.plotting.scatter_matrix(iris_dataframe, c=y_train, figsize=(15, 15),
                           marker='o', hist_kwds={'bins':20}, s=60, alpha=.8)
```

```python
#kNearest Neighbors Model using 15 neighbors
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=15)
```

```python
#Building the model
knn.fit(X_train, y_train)
```

```python
#Evaluating the model to check if prediction is accurate
import numpy as np
y_pred = knn.predict(X_test)
print('Test set predictions: ', y_pred)

#Checking model's accuracy using mean
print("Test set score: ", np.mean(y_pred==y_test))

#Checking model's accuracy using score
print('Test set score: ', knn.score(X_test, y_test))
```

```python
#New flower sample for prediction
import numpy as np
X_new=np.array([[4.5, 3.1, 1.5, 0.6]])
print("X_new shape: {}".format(X_new.shape))
```

```python
#Making the prediction of the new sample
prediction = knn.predict(X_new)
print("Prediction: ", prediction)
print("Predicted target name: ", iris_dataset['target_names'][prediction])
```
